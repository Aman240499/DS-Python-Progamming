```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.model_selection import train_test_split,KFold, cross_val_score
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 500
import seaborn as sns
from sklearn.metrics import r2_score,mean_squared_error,classification_report
%matplotlib inline
```

# Random Forest ***Regression**

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Boston.csv',
                index_col=0)
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.corr(),annot=True )
```

# ***Model Building***

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
rfr = RandomForestRegressor()
rfr.fit(xtrain,ytrain)
ypred = rfr.predict(xtest)
```

```python
print(f"R2 value: {r2_score(ytest,ypred)}") #accuracy
print(f"MSE: {mean_squared_error(ytest,ypred)}") #error
print(f"RMSE : {nm.sqrt(mean_squared_error(ytest,ypred))}")
```

```python
rfr.estimators_[0] #first decision tree
```

```python
len(rfr.estimators_)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(rfr.estimators_[0], feature_names=x.columns, filled=True, max_depth=2)
```

# Random Forest Classification

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Mobile_price_range.csv')
df
```

```python
df.describe()
```

```python
sns.countplot(x = df["price_range"])
```

```python
df["price_range"].replace([0,1,2,3],["low","medium","high","very High"],inplace=True)
```

```python
df
```

# **Model Building**

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain,ytrain)
ypred = rfc.predict(xtest)
```

```python
rfc.estimators_
```

```python
len(rfc.estimators_)
```

```python
print(classification_report(ytest,ypred))
```

```python
fig,index = plt.subplots(1,5, dpi = 1000)

for i in range(0,5):
    tree.plot_tree(rfc.estimators_[i],
                  feature_names=x.columns,
                  class_names=df['price_range'].unique(),
                   max_depth=2,
                   filled=True,
                  ax = index[i])
```
