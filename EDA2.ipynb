```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
%matplotlib inline
```

```python
#dummy dataset
df = pd.DataFrame({
    'Marks': [85,87,85,88,900],
    'Time' : [5,6,6.6,5,9.5]
})
df
```

```python
iso = IsolationForest(random_state=1)
iso.fit(df)
outlier = iso.predict(df)
outlier
```

```python
df["outliers"] = outlier
```
